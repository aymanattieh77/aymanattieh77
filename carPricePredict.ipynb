```python
# read data 
df = pd.read_csv('Car details v3.csv')
df
```

```python
# types of columns
df.dtypes
```

```python
# check null values 
missing_values = df.isnull().sum()
print(missing_values)
total_missing_value = (missing_values.sum()/df.size)*100
print('Total missing values in dataset : ', total_missing_value)
```

The cell above we have 222 null values in 5 columns, 
the Percentage of missing values is low so i will drop all rows that have null values *italicized text*

```python
print('the shape before drop null values ', df.shape)
df.dropna(inplace=True)
print('the shape after drop null values ', df.shape)
```

```python
from pandas.core.common import random_state
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

def dataset_score(x_train, x_test, y_train, y_test):
     model1 = RandomForestRegressor(n_estimators=100, random_state=99)
     model1.fit(x_train, y_train)
     predict1 = model1.predict(x_test)
     model2 = DecisionTreeRegressor()
     model2.fit(x_train, y_train)
     predict2 = model2.predict(x_test)   
     model3 = LinearRegression()
     model3.fit(x_train, y_train)
     predict3 = model3.predict(x_test)
     
     result1 = {'RandomForestRegressor (MAE):':mean_absolute_error(y_test, predict1),
               'DecisionTreeRegressor (MAE):':mean_absolute_error(y_test, predict2),
               'LinearRegression (MAE):':mean_absolute_error(y_test, predict3)}
     result2 = {'RandomForestRegressor (r2):':r2_score(y_test, predict1),
               'DecisionTreeRegressor (r2):':r2_score(y_test, predict2),
               'LinearRegression (r2):':r2_score(y_test, predict3)}          
     return result1, result2           
     
```

```python
# we need to covert categotical attributes to numeric attributes
# that will be done into method like Scaler encoder, but that will not preform well with columns have high number of unigue values greater than 15
# we need to drop columns with high cardinality
categorical_cols = df.select_dtypes(include='object')
high_cardinality = [cols for cols in categorical_cols if categorical_cols[cols].nunique()>10]

```

```python
# we can covert engine to numeric
new_df = df.drop(high_cardinality, axis=1)
```

```python
new_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
for i in new_df.select_dtypes(include='object'):
  new_df[i] = le.fit_transform(new_df[i])

new_df

```

```python
# define featurs and the target varible
X = new_df.drop('selling_price', axis=1)
y = new_df['selling_price']
```

```python
from sklearn.model_selection import train_test_split # for spliting the dataset into groups training and testing data

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=100)

result1, result2 = dataset_score(x_train, x_test, y_train, y_test)
print(result1)
print()
print(result2)
```
